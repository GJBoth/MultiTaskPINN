```python
from phimal_utilities.analysis import load_tensorboard
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
path_old = '../runs/Jul02_11-06-43_C02ZC3WHLVDM.local/'
path_new = '../runs/Jul02_10-55-02_C02ZC3WHLVDM.local/'
```

```python
df_old = load_tensorboard(path_old)
df_new = load_tensorboard(path_new)
```

```python
coeff_keys = [key for key in df_old.keys() if key[:len('coeff')] == 'coeff']
uncoeff_keys = [key for key in df_old.keys() if key[:len('unscaled')] == 'unscaled']
```

```python
df_old.keys()
```

```python
plt.plot(df_old.index, df_old['Total loss'], label='Old')
plt.plot(df_new.index, df_new['Total loss'], label='New')

plt.legend()
```

```python
plt.semilogy(df_old.index, df_old['MSE 0'], label='Old')
plt.semilogy(df_new.index, df_new['MSE 0'], label='New')

plt.legend()
```

```python
plt.semilogy(df_old.index, df_old['Regression 0'], label='Old')
plt.semilogy(df_new.index, df_new['Regression 0'], label='New')

plt.legend()
```

```python
plt.plot(df_old.index, df_old[coeff_keys])
plt.plot(df_old.index, df_old[coeff_keys[2]], linewidth=3)
plt.plot(df_old.index, df_old[coeff_keys[5]], linewidth=3)
```

```python
plt.plot(df_new.index, df_new[coeff_keys])
plt.plot(df_new.index, df_new[coeff_keys[2]], linewidth=3)
plt.plot(df_new.index, df_new[coeff_keys[5]], linewidth=3)
plt.xlim([0, 5000])
```

```python
plt.plot(df_new.index, np.abs(df_new[coeff_keys]))
plt.plot(df_new.index, np.abs(df_new[coeff_keys[2]]), linewidth=3)
plt.plot(df_new.index, np.abs(df_new[coeff_keys[5]]), linewidth=3)

plt.ylim([0, 4])
```

```python
plt.plot(df_new.index, df_new[uncoeff_keys])
plt.plot(df_new.index, df_new[uncoeff_keys[2]], linewidth=3)
plt.plot(df_new.index, df_new[uncoeff_keys[5]], linewidth=3)

plt.ylim([-1, 0.5])
```
